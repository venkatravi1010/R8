#### Convert the data to lower case for better matches

```python
# Converting all the data to lower case for better match
for columns in raw_data_1.columns:
    raw_data_1[columns] = raw_data_1[columns].str.lower() 

raw_data_1
```

```python
# Make a copy of the original dataframe
raw_data_2 = raw_data_1
raw_data_2 = raw_data_2.iloc[:,1:]
raw_data_2
# Find the unique keywords from the dataset
print(len(raw_data_2.stack().unique()))
```

```python
#### Logic to fill the adjacency matrix with the interaction strength
raw_data_1 = raw_data_1[raw_data_1['Title'].notna()]
raw_data_1 = raw_data_1[raw_data_1.notna()]
raw_data_1.reset_index(drop=True)

raw_data_1_clean = raw_data_1.drop(columns=['Title'])
trial_1 = raw_data_1_clean.stack()
df_trial = pd.DataFrame(trial_1)
words = df_trial[0].unique()

df_1 = raw_data_1
rows = len(df_1)

size = len(words)
adj_mat = np.zeros((size,size))

for row in range(0,rows):
  x = df_1.iloc[row,:]
  x = list(x)
  for i in range(0,size):
    for j in range(0,size):
      if i==j:
        pass
      elif words[i] in x:
        if words[j] in x:
          adj_mat[i,j]+=1

adj_mat = pd.DataFrame(adj_mat)
adj_mat.columns = words
adj_mat.index = words
adj_mat
```

```python
# All pairs of keywords with highest strength
top_10_tup = ()
for ut_r in words:
  max_value = 0
  temp = ()
  for ut_c in words:
    
    if adj_mat[ut_r][ut_c]>max_value:
      max_value = adj_mat[ut_r][ut_c]
      temp = (ut_r,ut_c,max_value,)
  top_10_tup = top_10_tup + temp
res_tup = tuple(top_10_tup[x:x+3] for x in range(0, len(top_10_tup), 3))

res_tup = list(res_tup)

def sort_tuple(tup):
    tup.sort(key = lambda x: x[2], reverse=True)
    return tup

res = sort_tuple(res_tup)
print(res[0:11])
```

```python
# To compute node degree and printing top 10 nodes
from heapq import nlargest

node_degree = {}

for ut in words:
  degree = 0
  for i in range(adj_mat.shape[1]):
    if int(adj_mat.loc[:,ut][i])>0:
      degree = degree+1
  node_degree.update({ut:degree})

top_10 = nlargest(10, node_degree, key = node_degree.get)
for v in top_10:
  print(v, ":", node_degree.get(v))

```

```python
 # To compute node strength
from heapq import nlargest


node_strength = {}
for ut in words:
  node_strength.update({ut:(adj_mat[ut].sum())/node_degree.get(ut)})

top_10 = nlargest(10, node_strength, key = node_strength.get)
for v in top_10:
  print(v, ":", node_strength.get(v))
  
```

```python
# Visualisation of Node Degree vs Avg. Strength
import matplotlib.pyplot as plt
sns.lineplot(x=node_degree,y=node_strength)
plt.xlabel('Node Degree')
plt.ylabel('Avg. Strength')
plt.show()
```
